```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
rv=sts.rayleigh(np.random.randint(10))
sample=rv.rvs(size=1000)

x=np.linspace(0,15,100)
pdf=rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()

n=np.random.randint(100)
samples_count=1000
a1=np.array([])

for number in range(samples_count):
    sample=rv.rvs(size=n)
    a1=np.append(a1, sample.mean())

label_1='sample of ' + str(n) + ' histogram'
plt.hist(a1, density=True, label=label_1)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```
