```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
grid = np.linspace(-7, 20, 1000)  #    
loc_values = [0, 3, 0, 5, 6]  #    a
sigma_values = [1, 1, 2, 8, 0.5]  #    sigma

plt.figure(figsize=(12, 6)) 

for i, (a, sigma) in enumerate(zip(loc_values, sigma_values)):
    plt.plot(grid, sps.norm(a, sigma).pdf(grid), lw=5, 
             label='$\mathcal{N}' + '({}, {})$'.format(a, sigma))

plt.legend(fontsize=16)
plt.title('  ', fontsize=20)
plt.xlabel('', fontsize=18)
plt.ylabel('', fontsize=18)
plt.grid(ls=':')
plt.show()
```

```python
grid = np.linspace(-1, 10, 1000) 
lam_values = [1, 4, 2, 8, 0.5]

plt.figure(figsize=(12, 6)) 

for i, (lam, lam) in enumerate(zip(lam_values, lam_values)):
    plt.plot(grid, sps.expon(lam, lam).pdf(grid), lw=5, 
             label='$\mathcal{L}' + '({})$'.format(lam))

plt.legend(fontsize=16)
plt.title('  ', fontsize=20)
plt.xlabel('', fontsize=18)
plt.ylabel('', fontsize=18)
plt.grid(ls=':')
plt.show()
```

```python
grid = np.linspace(-5, 5, 1000) 
n_values = [1, 4, 2, 8, 0.5]

plt.figure(figsize=(12, 6)) 

for i, n in enumerate(n_values):
    plt.plot(grid, sps.t(n).pdf(grid), lw=5, 
             label='$\mathcal{n} ' + '({})$'.format(n))

plt.legend(fontsize=16)
plt.title('  ', fontsize=20)
plt.xlabel('', fontsize=18)
plt.ylabel('', fontsize=18)
plt.grid(ls=':')
plt.show()
```
